```python
# 2번
df_last=pd.read_csv('C:/ai/downloads/shareData/부동산/주택도시보증공사_전국 신규 민간아파트 분양가격 동향_20241231.csv',encoding='cp949')
df_last
```

```python
def commadrop(row):
    row = row.copy()
    price = row['분양가격(제곱미터)']
    if price is np.nan:
        pass
    elif price.strip() == '': 
        row['분양가격(제곱미터)'] = np.nan
    elif price.find(',') != -1:
        pricesplited = price.split(',')
        row['분양가격(제곱미터)'] = float(''.join(pricesplited))
    else:
        row['분양가격(제곱미터)'] = float(price)
    return row
```

```python
df_last.apply(commadrop, axis=1)
```

```python
df_last['분양가격(제곱미터)'] = pd.to_numeric(df_last['분양가격(제곱미터)'].str.strip().str.replace(',',''))
df_last['평당분양가격'] = df_last['분양가격(제곱미터)']*3.3
df_last.head()
```

```python
#3번
fillnavalue=df_last.pivot_table(index='연도', columns='지역명', values='평당분양가격')
fillnavalue
```

```python
#4번
sns.boxplot(data=df_last, x='연도', y='평당분양가격')
plt.show()
```

```python
# 5번
new_first = df_first.melt(id_vars='지역', 
                          var_name='연도월', 
                          value_name='평당분양가격')
new_first['연도'] = new_first['연도월'].map(lambda datestr : int(datestr.split('년')[0])
new_first['월'] = new_first['연도월'].apply(lambda datestr : int(datestr.split('년')[-1][:-1]))
new_first.drop('연도월',axis=1,inplace=True)
new_first =new_first.rename(columns={'지역':'지역명'})
new_first
```

```python
import re
df_last['전용면적'] = df_last['규모구분'].apply(lambda x : re.sub('전용면적|이하| ', '', x).replace('제곱미터', 'm²').replace('초과','~'))
df_last.drop(['규모구분','분양가격(제곱미터)','전용면적'],axis=1, inplace=True)
df_last
```

```python
col = df_last.pop('평당분양가격') # '전용면적'컬럼 빼서 col변수 할당
df_last.insert(1, '평당분양가격', col)
df_last
```

```python
df=pd.concat([new_first,df_last]).reset_index(drop=True)
df
```
